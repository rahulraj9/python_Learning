# Indexing with Boolean Arrays
### Boolean arrays are useful when you need to select a dataset according to set criteria.





* Here, the original dataset contains test scores of two students. You can use a Boolean array to choose only the scores that are above a given value.


```python
test_score = np.array([[83,71,57,63],[54,68,81,45]])
print(test_score)
```

```python
#set passing score
passing_score = test_score>60
print(passing_score)
```

```python
# Send passing score as an argument to test scores object
test_score[passing_score]
```

# Copy and Views
* When working with arrays, data is copied into new arrays only in some cases. Following are the three possible scenarios* :


```python
#In this method, a variable is directly assigned the value of another variable. No new copy is made
cafe_near_me = np.array(['chai e dil','coffee point','Friends cafe','costa cofee']) #original data_set
print(cafe_near_me)
```

```python
near_me_cafe = cafe_near_me #assigned data_set
print(near_me_cafe)
```

```python
cafe_near_me is near_me_cafe
```

```python
# A view, also referred to as a shallow copy, creates a new array object.
cafe_near_me   #original data_set
```

```python
view_of_cafe_near_me = cafe_near_me.view()
```

```python
len(view_of_cafe_near_me)
```

```python
view_of_cafe_near_me[3]='Data_Changed'  #Data_Changed_in_view_object
```

```python
view_of_cafe_near_me
```

```python
cafe_near_me  #original Data_Set also changed
```

```python
#Copy is also called deep copy because it entirely copies the original dataset. Any change in the copy will not affect the original dataset.
cafe_near_me #original data_set
```

```python
copy_of_cafe_near_me = cafe_near_me.copy()
print(copy_of_cafe_near_me)
```

```python
copy_of_cafe_near_me is cafe_near_me  #Shows copy and original object are different
```

```python
len(copy_of_cafe_near_me)
```

```python
copy_of_cafe_near_me[3] = 'hello'
```

```python
print(copy_of_cafe_near_me)  #original data_set is  not changed
print(cafe_near_me)
```
